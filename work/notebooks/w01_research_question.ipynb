## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The decision this work aims to improve is:

"Which content pages should a search/content team review first for possible refresh opportunities?"

The people acting on this output would be content strategists, SEO specialists, or reviewers who decide whether a page should be refreshed, expanded, monitored, or left unchanged.

A wrong recommendation has a cost because review time is limited. A false positive may waste effort on pages that do not need attention, while a false negative may cause valuable pages with declining visibility or missed opportunities to be overlooked. Therefore, the goal is to improve prioritization rather than make fully automated decisions.
