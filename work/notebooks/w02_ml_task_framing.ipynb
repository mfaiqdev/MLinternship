## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The ideal target would be whether a page **actually improves after a content refresh**. However, the starter dataset does not contain confirmed refresh outcomes.

Instead, I will use **`trend_direction = "down"`** as a **proxy target**. This label is created from observed performance trends and serves as an indicator that a page may deserve review.

This is a **defined rule based on observed data**, not a proven future outcome. The goal is to identify pages that are more likely to require attention while recognizing that the proxy is not a guarantee of future performance.

```python
proxy_target = (df["trend_direction"] == "down").astype(int)

print("Proxy target: trend_direction == down")
print("Positive rows:", proxy_target.sum())
print("Positive rate:", round(proxy_target.mean(), 3))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can identify simple cases, such as pages with declining traffic or old content that has not been updated recently. However, real content performance depends on **multiple interacting signals** rather than a single condition.

For example, one page may have high impressions but low CTR, while another may have moderate traffic but rapidly declining engagement. These combinations are difficult to capture using a small set of if-statements.

Machine learning can learn patterns from many observed signals at the same time and produce a more useful ranking for human review. The final output remains **decision-support**, and every recommendation should still be reviewed by a person before any action is taken.
